## Regulation
Notice that in the last reaction (PDH) the line

NREGULATION ACETYL-COA 1.00e-03 20.0

The keyword NREGULATION is for negative regulation using a Hill Function with constant = 6.1e-04 and an exponent of 20.0
$$Activity = \frac{K^{n}}{K^{n} + X^{n}}$$

where K is the constant similar to the Michaelis constant, n is the exponent that controls the steepness of the Hill function and $X$ is the concentration of the species that regulates the activity. The Hill function is used to update the activity of the enzyme as follows:

```python
constant = 6.1e-04
exponent = 10
constant_pow_exponent = np.exp(exponent*np.log(constant))

concentration = np.arange(1.0e-4, 1.0e-3, 1.0e-5)
concentration_pow_exponent = np.exp(np.multiply(exponent,np.log(concentration)))

activity = np.divide(constant_pow_exponent,(constant_pow_exponent + concentration_pow_exponent))
plt.plot(concentration,activity)
```

Likewise, positive regulation is done using the comparable Hill Function,
$$Activity = \frac{X^{n}}{K^{n} + X^{n}}$$.
Positive regulation is indicated in the Boltzmann .dat file with keyword PREGULATION.

```python
constant = 6.1e-04
exponent = 10
constant_pow_exponent = np.exp(exponent*np.log(constant))

concentration = np.arange(1.0e-4, 1.0e-3, 1.0e-5)
concentration_pow_exponent = np.exp(np.multiply(exponent,np.log(concentration)))

activity = np.divide(concentration_pow_exponent,(constant_pow_exponent + concentration_pow_exponent))
plt.plot(concentration,activity)
```

# Integration with Experimental Data


For the following discussion, consider the enzymatic reaction involving the conversion of metabolite $A$ to metabolite $B$ catalyzed by enzyme $E$,
$$A + E \longrightarrow AE \longrightarrow BE \longrightarrow B + E$$

The total free energy change for the reaction is 
$$\Delta G_{T} = \Delta G_{1} + \Delta G_{2} + \Delta G_{3}$$.
where, 
$$
\Delta G_{1} = -RT log\frac{[A][E]}{[AE]}\\
e^{-\Delta G_{1}/RT} = \frac{[A][E]}{[AE]} \\
$$
$$
\Delta G_{2} = -RT log\frac{[AE]}{[BE]} \\
e^{-\Delta G_{2}/RT} = \frac{[AE]}{[BE]} \\
$$

$$
\Delta G_{3} = -RT log\frac{[BE]}{[B][E]} \\
e^{-\Delta G_{3}/RT} = \frac{[BE]}{[B][E]} \\
$$

Assuming maximum entropy (and therefore maximum efficiency) conditions, we have $\Delta G_{1} = \Delta G_{2} = \Delta G_{3} = \frac{1}{3} \Delta G_{T}$.

## Prediction of total population of each metabolite from simulation
The total population of each metabolite consists of the population of the metabolite that is free in solution in the cytosol and the population of the metabolite that is bound to enzymes. This prediction carries the caveat that we are modeling all of the proteins that bind a particular metabolite. That is, the predicted value of the total population of a metabolite assumes that we are including in the analysis all proteins that bind the metabolite, because the total population of a metabolite consists of both the unbound $[A]$ and bound populations $[EA]$.

The total concentration of metabolite $A$ is, 
$$
[A] + [AE] = [A] + [A][E]e^{\Delta G_{1}/RT} \\
=[A] (1 + [E]e^{\Delta G_{1}/RT})
$$

## Predict Enzyme Levels and Saturation from Simulated and Measured Metabolites.
Solving for the enzyme concentration $[E]$,
$$
[E]= (\frac{[A] + [AE]}{[A]} -1)e^{-\Delta G_{1}/RT}.
$$
The total concentration of $A$, $[A] + [AE]$, is available from the prediction just described or from experimental measurement while the concentration of free $A$, $[A]$, can be predicted directly from the simulation. At maximum entropy (and maximum efficiency), each reaction is equally likely such that $\Delta G_{1} = \Delta G_{2} = \Delta G_{3} = \frac{1}{3} \Delta G_{T}$. The free energy $\Delta G_{T}$ can be predicted from simulation or calculated from metabolic flux analysis data if information can be obtained for both forward and reverse rates for the reaction. These predicted enzyme concentrations can then be compared to measured enzyme concentrations, if available, and enzyme regulation can be inferred by analyzing the difference between enzyme activity and enzyme concentration.

Likewise, the percent saturation of enzymes can be predicted,

$$
\frac{[AE] + [BE]}{[E]} = [A]e^{\Delta G_{1}/RT} + [B]e^{-\Delta G_{3}/RT}
$$

## Prediction of free population of metabolites from MFA and absolute concentration measurements
Data from metabolic flux analysis (MFA) can be used to predict the population of metabolites that are free in solution. Here we assume the maximum entropy condition again such that $\Delta G_{1} = \frac{1}{3}\Delta G_{T}$ and that both forward $J_{+}$ and reverse flux $J_{-}$ values for a reaction can be measured by MFA. At steady state, $\Delta G_{T} = -RTlog\frac{J_{+}}{J_{-}}$. Then if the total measured population of metabolite $A$ is $[A] + [AE]$, the concentration of the free metabolite population is given by, 

$$
[A] = \frac{[A] + [AE]}{(1 + [E]e^{\Delta G_{1}/RT})}. 
$$
This calculation depends on knowing the population of the respective enzyme that is free (unbound by $A$) in solution.
